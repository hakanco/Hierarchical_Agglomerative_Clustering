```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

### Creating a blob of 200 data points

```python
dataset = make_blobs(n_samples=500,
                    n_features=2,
                    centers=4,
                    cluster_std=1.9,
                    random_state=0)
```

```python
points = dataset[0]
```

```python
targets = dataset[1]
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

### Creating a dendrogram

```python
dendrogram = sch.dendrogram(sch.linkage(points, method='ward'))
```

```python
plt.scatter(dataset[0][:,0], dataset[0][:,1])
```

```python
hc = AgglomerativeClustering(n_clusters=4,
                             affinity='euclidean', linkage='ward')
```

```python
y_hc = hc.fit_predict(points)
```

```python
plt.scatter(points[y_hc == 0, 0], points[y_hc == 0, 1], s=100, c='blue')
plt.scatter(points[y_hc == 1, 0], points[y_hc == 1, 1], s=100, c='yellow')
plt.scatter(points[y_hc == 2, 0], points[y_hc == 2, 1], s=100, c='red')
plt.scatter(points[y_hc == 3, 0], points[y_hc == 3, 1], s=100, c='orange')
```
